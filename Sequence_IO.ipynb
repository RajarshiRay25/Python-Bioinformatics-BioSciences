```python
# import libraries
from Bio import SeqIO
```

```python
# parse files from orchid.fasta
## SeqIO.read --> for single file entry
## SeqIO.parse --> for multi file entries

for fastaRecord in SeqIO.parse("./orchid.fasta","fasta"):
    print(fastaRecord.id)
    print(fastaRecord.description)
    print(fastaRecord.seq)
print(f"Total Records :{len(fastaRecord)} ")
```

```python
# get all ids of organisms

names =[]
for orgName in SeqIO.parse("./orchid.fasta","fasta"):
    names.append(orgName.id)

names
```
