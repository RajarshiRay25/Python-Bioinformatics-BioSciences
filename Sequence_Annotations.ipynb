```python
# import libraries
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
from Bio import SeqIO
```

```python
# input sequence

mySeq = Seq("ATGCGATAGCGAGATCGA")

# record the sequence for annotations

mySeqRecord = SeqRecord(mySeq)
mySeqRecord
```

```python
# Now lets put certain parameters

mySeqRecord.seq   # return sequence
mySeqRecord.id = "CA1234"  # set a id/accession identifier
mySeqRecord.description = "This is a test data"   # set a description
mySeqRecord.name = "Test Data DNA"

# View the final data object
print(mySeqRecord)
```

```python
# taking input from fna/fasta file

fastaRecord = SeqIO.read("./testfile.fna","fasta")
fastaRecord
```

```python
# Access properties from fasta
print(fastaRecord.name)
print(fastaRecord.id)
print(fastaRecord.description)
print(fastaRecord.seq)
```

```python
# Access genbank file format -- NCBI
genbankRecord = SeqIO.read("./testgenbank.gb","genbank")
genbankRecord
```
